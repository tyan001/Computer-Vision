```python
from math import ceil, exp, pi, log, cos, sin
import matplotlib.pyplot as plt
from scipy import ndimage
import numpy as np
import glob
from skimage import color, io
from skimage.transform import resize
import sys
from numpy import array, zeros
```

# Introduction
**In this lab we will translate images to their log-polar plot form. We will explore how the log polar plot image changed after a transformation is operated it on it like rotation**

# Methodolgy
The algorithm to create the log polar plot is by: <br>
1. set a coordinate as the the center point. <br>
2. Translate the input image to a new graph, where the x axis corresponds to the log of the distance from the center point and y axis correspond to the angle of th distance.


Here are the equations: where x_{c}<br>
$ r = e^{log(p)} = \sqrt{(x - x _c)^2 + (y - y _c)^2}$<br>
$ x = r cos(\theta)  $<br>
$ y = r sin(\theta)  $

```python
ima = []
for filename in glob.glob('../imagesHW0/*.jpg'): #assuming gif
    im = color.rgb2gray(io.imread(filename))
    im = resize(im, (512, 512), anti_aliasing=True)
    ima.append(im)
```

```python
def show(o_img, log_0, log_90,log_180):
    fig = plt.figure()
    fig.set_figheight(25)
    fig.set_figwidth(25)
    fig.add_subplot(2,2,1)
    plt.imshow(o_img, cmap='gray')
    # display the new image
    fig.add_subplot(2,2,2)
    plt.imshow(log_0, cmap='gray')
    fig.add_subplot(2,2,3)
    plt.imshow(log_90, cmap='gray')
    fig.add_subplot(2,2,4)
    plt.imshow(log_180, cmap='gray')
    plt.show(block=True)
```

```python
def showRotatePolar(image):
    log_polar_0 = []
    log_polar_90 = []
    log_polar_180 = []
    for x in image:
        rotate90 = ndimage.rotate(x,90)
        rotate180 = ndimage.rotate(x,180)
        
        log_polar = polarTrans(x)
        log_polar90 = polarTrans(rotate90)
        log_polar180 = polarTrans(rotate180)
        
        log_polar_0.append(log_polar)
        log_polar_90.append(log_polar90)
        log_polar_180.append(log_polar180)
        
        show(x,log_polar,log_polar90,log_polar180)
        
    return log_polar_0,log_polar_90,log_polar_180
```

```python
_transforms = {}
def _get_transform(i_0, j_0, i_n, j_n, p_n, t_n, p_s, t_s):
    transform = _transforms.get((i_0, j_0, i_n, j_n, p_n, t_n))
    
    if transform == None:
        i_k = []
        j_k = []
        p_k = []
        t_k = []

        for p in range(0, p_n):
            p_exp = exp(p * p_s)
            for t in range(0, t_n):
                t_rad = t * t_s

                i = int(i_0 + p_exp * sin(t_rad))
                j = int(j_0 + p_exp * cos(t_rad))

                if 0 <= i < i_n and 0 <= j < j_n:
                    i_k.append(i)
                    j_k.append(j)
                    p_k.append(p)
                    t_k.append(t)

        transform = ((array(p_k), array(t_k)), (array(i_k), array(j_k)))
        print(transform)
        _transforms[i_0, j_0, i_n, j_n, p_n, t_n] = transform

    return transform

def logpolar(image, p_n=None, t_n=None):
    i_0 = image.shape[0]/2
    j_0 = image.shape[1]/2
    (i_n, j_n) = image.shape[:2]
    
    i_c = max(i_0, i_n - i_0)
    j_c = max(j_0, j_n - j_0)
    d_c = (i_c ** 2 + j_c ** 2) ** 0.5
    
    if p_n == None:
        p_n = int(ceil(d_c))
    
    if t_n == None:
        t_n = j_n
    
    p_s = log(d_c) / p_n
    t_s = 2.0 * pi / t_n
    
    (pt, ij) = _get_transform(i_0, j_0, i_n, j_n, p_n, t_n, p_s, t_s)

    transformed = zeros((p_n, t_n) + image.shape[2:], dtype=image.dtype)

    #print(pt)
    #print(ij)
    
    transformed[pt] = image[ij]
    return transformed
```

```python
trans_list = {}
def polTranslate(h_mid,w_mid, h, w, rtop, t_n, p, t):
    transform = trans_list.get((h_mid,w_mid, h, w, rtop, t_n))
    
    if transform == None:
        i_k = []
        j_k = []
        p_k = []
        t_k = []

        for l in range(0, rtop):
            l_exp = exp(l * p)
            for q in range(0, t_n):
                t_rad = q * t

                i = int(h_mid + l_exp * sin(t_rad))
                j = int(w_mid + l_exp * cos(t_rad))

                if 0 <= i < h and 0 <= j < w:
                    i_k.append(i)
                    j_k.append(j)
                    p_k.append(p)
                    t_k.append(t)

        transform = ((array(p_k), array(t_k)), (array(i_k), array(j_k)))
        print(transform)
        trans_list[h_mid,w_mid, h, w, rtop, t_n] = transform
    return transform
```

```python
def polarTrans(image):
    h,w = image.shape
    h_mid = h / 2
    w_mid = w / 2
    
    h_center = max(h_mid, h-h_mid)
    w_center = max(w_mid, w-w_mid)
    
    r = (h_center ** 2 + w_center ** 2) ** 0.5 
    rtop = int(ceil(r))
    
    t_n = w
    
    p = log(r) / rtop
    t = 2.0 * pi / w
    
    (point, hw) = polTranslate(h_mid,w_mid, h, w, rtop, t_n, p, t)
    #print(point)
    #print(hw)
    transformed = zeros((rtop, t_n) + image.shape[2:], dtype=image.dtype)   
    transformed[point] = image[hw]
    
    return transformed
```

```python
def find_angle(img_og_log,img_rot_log):
    minimum = sys.maxsize
    index = 0
    for i in range(img_rot_log.shape[0]):
        a = (abs(img_og_log[0] - np.roll(img_rot_log[0],i)).sum())
        if(a < minimum):
            minimum = a
            index = i
    #print(img_rot_log.shape[1]-index)
    #print(img_rot_log.shape[1])
    angle = 360*(index)/(img_rot_log.shape[1])
    return angle
```
